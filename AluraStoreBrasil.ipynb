### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja2.head()
```

# 4. Frete Médio por Loja + Plotagem

```python
import pandas as pd
import matplotlib.pyplot as plt


tipo1 = loja.groupby('Categoria do Produto')

resultado1 = tipo1.agg({
    'Categoria do Produto': 'count',
    'Avaliação da compra': 'mean',
    'Frete': 'mean'
})

r_rename = resultado1.rename(columns={
    'Categoria do Produto': 'Qtd_Vendas',
    'Avaliação da compra': 'Nota_Media', 
    'Frete': 'Frete_media'
})

# Criação de bloco com 1 linha e 3 colunas
figsize=(18, 6)
fig, ax = plt.subplots(1, 3, figsize=(18, 6))


# --- GRÁFICO 1: Vendas e Frete (Barras) 

r_rename[['Qtd_Vendas', 'Frete_media']].plot(kind='bar',
    ax=ax[0], 
    color=['skyblue', 'salmon'],
    ylabel='')
ax[0].set_title("Vendas vs. Frete Médio")
ax[0].set_ylabel("Valores")

# --- GRÁFICO 2: Proporção de Vendas (Pizza) ---

r_rename['Qtd_Vendas'].plot(
    kind='pie', 
    ax=ax[1], 
    autopct='%1.1f%%', 
    legend=True,    
    ylabel=''          # Remove o rótulo lateral feio (Qtd_vendas)
)
ax[1].set_title("Vendas por Categoria")
ax[1].legend(loc='center left', bbox_to_anchor=(0.9, 0.5)) # Legenda deslocada




# --- GRÁFICO 3: Nota Média (Barras Horizontais) ---

r_rename['Nota_Media'].plot(kind='barh', 
    ax=ax[2], 
    color='lightgreen',
    ylabel=''
    )
ax[2].set_title("Nota Média de Avaliação")
ax[2].set_xlim(0, 5) # Fixamos o eixo X de 0 a 5 (nota máxima)


# ajuste para dar espaço
plt.tight_layout()
plt.show()
```
